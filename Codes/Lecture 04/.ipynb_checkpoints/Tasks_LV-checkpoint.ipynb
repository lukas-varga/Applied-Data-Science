# Task 1

In: bill_of_rights.txt

Out: stats.txt

```python
r = open("bill_of_rights.txt", 'r')
words_num=0 
rows_num=0
counts = {}
max_word="each word has only 1 occurrence"
max_occ=0
for line in r:
    #make every letter small and delete white spaces
    line = line.lower()
    line = line.replace("\r","")
    line = line.replace("\n","")
    line = line.replace("\r\n","")
    line = line.replace("\t","")
    #in whole text there is as many words as spaces +1 for each row
    for i in line:
        if i == " ":
            words_num += 1
    words_num += 1
    
    #number of rows
    rows_num += 1
    
    #split the line by symbol space in order to count them
    words = line.split(" ")
    #print(words)
    for j in words:
        if j not in counts:
            counts[j]=1
        else:
            counts[j]+=1
            #keep most frequent word in memory
            if(counts[j]>max_occ):
                max_occ = counts[j]
                max_word = j
r.close()   

w = open("stats.txt",'w')
w.write("Number of rows: "+str(rows_num)+"\n")
w.write("Number of words: "+str(words_num)+"\n")
w.write("Most frequent word: '"+str(max_word)+"' ("+str(max_occ)+" times)"+"\n")
w.close()
print("Done")
```

# Task 2

 In: gps.csv

```python
import numpy as np
import pandas as pd
import time
import datetime

#convert to unix time
def computeUnix(x):
    return time.mktime(datetime.datetime.strptime(x,"%Y/%m/%d-%H:%M:%S").timetuple())

#time->UTC; ltime->USA time
df = pd.read_csv("gps.csv")
df=df.rename(columns={"long": "lon"})
df["unix timestamp"] = df["time"].apply(lambda x: computeUnix(x))
df[["lat","lon","unix timestamp"]]
```

# Task 3

 In: test_assignment_data.csv

```python
import numpy as np
import pandas as pd

def gdpPerCap(row):
    return row["Total GDP"]/row["Population"]

df=pd.read_csv("test_assignment_data.csv")
df=df.rename(columns={"country": "Country","POP": "Population","tcgdp": "Total GDP"})
df["GDP Percap"] = df.apply(lambda row: gdpPerCap(row),axis=1)
df[["Country","Population","Total GDP","GDP Percap"]]
```

1. Find the month and day where “rain” measurement is NOT Zero. (2pts)

```python
import pandas as pd
import numpy as np

df=pd.read_csv("forestfires.csv")
df = df[df["rain"]!=0]
df[["month","day","rain"]]
```

2. Find the descriptive statistics of “temp” and “wind” (2pts)

```python
import pandas as pd
import numpy as np

df=pd.read_csv("forestfires.csv")
raw_data = {"temp": df["temp"].describe(),
           "wind": df["wind"].describe()}
stats = pd.DataFrame(raw_data,columns=["temp","wind"])
stats
```

3. Find mean temperature for each month and plot it (2pts)

```python
import pandas as pd
import numpy as np

df=pd.read_csv("forestfires.csv")
by_month = df.groupby("month")
res = by_month.temp.mean()
print(res)
```

```python
res.plot()
```

4. Find the month and day with the highest and lowest temperature. (2pts)

```python
import pandas as pd
import numpy as np

df=pd.read_csv("forestfires.csv")
high = df[df["temp"] == df["temp"].max()]
high[["month","day","temp"]]
```

```python
low = df[df["temp"] == df["temp"].min()]
low[["month","day","temp"]]
```

5. Find the mean temperature at different unique locations (x,y) (2pts) 

```python
import pandas as pd
import numpy as np

df=pd.read_csv("forestfires.csv")
by_xy = df.groupby(["X","Y"])
res = by_xy.temp.mean()
print(res)
```
