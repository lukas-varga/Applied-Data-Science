## curve_fit

```python
import numpy as np
import matplotlib . pyplot as plt
# Lets say we have following data
xdata = np.array([-2,-1.64,-1.33,-0.7,0,0.45,1.2,1.64,2.32,2.9])
ydata = np.array([0.699369,0.700462,0.695354,1.03905,1.97389,2.41143,1.91091,0.919576, -0.730975,-1.42001])

plt.plot(xdata,ydata,"*")

```

```python
from scipy.optimize import curve_fit
# define the function

def func(x, p1,p2):
    return p1*np.cos(p2*x) + p2*np.sin(p1*x)

# guess parameters Use a starting guess
#p1=1
#p2=0.2

# fit parameters
popt, pcov = curve_fit(func, xdata, ydata, p0=(1.0,0.2))
print(popt) # this variable stores parameter values

p1 = popt[0]
p2 = popt[1]
residuals = ydata - func(xdata,p1,p2)
res = sum(residuals**2)
print(res)

curvex = np.linspace(-2,3,100) # create more data points in between
survey = func(curvex,p1,p2)

plt.plot(survey)
plt.xlabel('xdata')
plt.ylabel('ydata')
plt.show()
```

```python
import numpy as np
import matplotlib . pyplot as plt
from scipy.optimize import curve_fit

# define the function
def func(x, p1,p2):
    return p1*np.cos(p2*x) + p2*np.sin(p1*x)

# Lets say we have following data
xdata = np.array([-2,-1.64,-1.33,-0.7,0,0.45,1.2,1.64,2.32,2.9])
ydata = np.array([0.699369,0.700462,0.695354,1.03905,1.97389,2.41143,1.91091,0.919576, -0.730975,-1.42001])


# guess parameters Use a starting guess
#p1=1
#p2=0.2

# fit parameters
popt, pcov = curve_fit(func, xdata, ydata, p0=(1.0,0.2))
print(popt) # this variable stores parameter values

p1 = popt[0]
p2 = popt[1]
residuals = ydata - func(xdata,p1,p2)
res = sum(residuals**2)
print(res)

curvex = np.linspace(-2,3,100) # create more data points in between
curvey = func(curvex,p1,p2)

plt.plot(xdata,ydata,"*")
plt.plot(curvex,curvey)
plt.xlabel('xdata')
plt.ylabel('ydata')
plt.show()
```

## interpolate

```python
import scipy.interpolate as sp
# 50 real points of sin(x) in [0 10]
xx = np.linspace(0, 10, 50)
yy = np.sin(xx)

# 10 sample of sin(x) in [0 10]
x = np.linspace(0, 10, 10)
y = np.sin(x)

# interpolation
fl = sp.interp1d(x, y,kind='linear')
fc = sp.interp1d(x, y,kind='cubic')

# create sample data
xnew = np.linspace(0, 10, 50)
# interpolation 
y_il = fl(xnew) 
y_ic = fc(xnew)

plt.plot(x,y,"*", label="sample")
plt.plot(xnew, y_ic,label="cubic")
plt.plot(xnew, y_il,label="linear")
plt.legend(loc='upper left')
plt.xlabel("xdata")
plt.ylabel("ydata")
```

## quad

```python
from scipy.integrate import quad
import numpy as np

# Defining function to integrate

func = lambda x: np.cos(np.exp(x))**2
#integrating function with uppper and lower
#limits of 0 to 3, respectively
solution=quad(func,0,3)

print(solution)

# The first element is the desired value
# and the second is the error                   
```

## trapz

```python
from scipy.integrate import trapz

x = np.sort(np.random.randn(150)*4+ 4).clip(0,5) 
func1 = lambda x: np.sin(x) * np.cos(x ** 2) + 1
y = func1(x)

dsolution = trapz(y, x=x)

print(dsolution)
```

## kmeans and vq

```python
from scipy.cluster.vq import kmeans,vq

# data generation

data = np.vstack((np.random.rand(150,2) + np.array([.5,.5]),np.random.rand(150,2)))

# computing K-Means with K = 2 (2 clusters)
centroids,distortion = kmeans(data,2)



# assign each sample to a cluster
idx,distortion1 = vq(data,centroids)



# # some plotting using numpy's logical indexing
plt.plot(data[idx==0,0],data[idx==0,1],'ob',data[idx==1,0] ,data[idx==1,1],'or')
plt.plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
```

## Fast Fourier Transform

Fourier analyis a method for expressing a function as a sum periodic componets. This example demonstrates a time wave and applying a Discrete Fourier Transform 

```python
from scipy import fftpack
f = 10  # Frequency, in cycles per second, or Hertz
f_s = 100  # Sampling rate, or number of measurements per second

t = np.linspace(0, 2, 2 * f_s, endpoint=False)
x = np.sin(f * 2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, x)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Signal amplitude');


X = fftpack.fft(x)
freqs = fftpack.fftfreq(len(x)) * f_s

fig, ax = plt.subplots()

ax.stem(freqs, np.abs(X))
ax.set_xlabel('Frequency in Hertz [Hz]')
ax.set_ylabel('Frequency Domain (Spectrum) Magnitude')
ax.set_xlim(-f_s / 2, f_s / 2)
ax.set_ylim(-5, 110)
```

```python
import scipy.misc
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt

face = scipy.misc.face(gray=True)

lx, ly = face.shape

X, Y = np.ogrid[0:lx, 0:ly]

mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/4

face[mask] = 0
plt.imshow(face, cmap=plt.cm.gray)
```
