## Exercise 1: Read Total profit of all months and show it using a line plot

In this activity your will provide the total profit data provided for each month. The generated line plot must include the following properties: â€“

X label name = Month Number
Y label name = Total profit
The line plot graph should look like this.

![Image 1](images/e1.png)



```python
# Excercise 1

#Start Code 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/company_sales_data.csv")
x = data["month_number"]
y = data["total_profit"]
plt.plot(x,y)

#labels
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.title("Company profit per month")

#limits and ticks
plt.xticks(x)
plt.yticks(np.arange(100000, 500001, 100000))
plt.show()

#end code
```

## Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.

- Line width should be 3

The line plot graph should look like this.

```python
# Excercise 2

#Start Code 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/company_sales_data.csv")
x = data["month_number"]
y = data["total_profit"]

#formatting curve
plt.plot(x,y,'--o',color="red",markerfacecolor="black",linewidth=3.0,label = "Profit data of last year")

#labels
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.title("Company Sales data of last year")

#limits and ticks
plt.xticks(x)
plt.yticks(np.arange(100000, 500001, 100000))

#legend
plt.legend(loc="lower right")
plt.show()

#end code
```

## Exercise Question 8: Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

![Image 8](images/e8.png)

```python
# Excercise 8

#Start Code 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#read data
df = pd.read_csv("data/company_sales_data.csv")
data=[]
data.append(df["facecream"])
data.append(df["facewash"])
data.append(df["toothpaste"])
data.append(df["bathingsoap"])
data.append(df["shampoo"])
data.append(df["moisturizer"])

#sum units of all kinds of gods
sums=[]
for i in data:
    sums.append(i.sum())

#total number of units
total=0
for i in range(len(sums)):
    total+=sums[i]

#make percentage
perc=[]
for i in sums:
    result = i/total
    perc.append(result)
    
#make dictionary
res={}
for i in range(len(data)):
    res[data[i].name] = perc[i]
    
#plot
plt.pie(list(res.values()),labels=res.keys(), autopct='%1.1f%%', shadow=False)

#labels
plt.title("Sales Data")

#legend
plt.legend(loc="lower right",bbox_to_anchor=(1.25, 0))
plt.show()

#end code
```

## Exercise Question 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
# Excercise 9

#Start Code 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#data
df = pd.read_csv("data/company_sales_data.csv")
x = df["month_number"]
t1=df["bathingsoap"]
t2=df["facewash"]

#plots
fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.plot(x,t1,"-o",color="black",linewidth=2.0)
ax1.set_title("Sales data of a Bathingsoap")
ax1.set_xticks([])

ax2 = fig.add_subplot(212)
ax2.plot(x,t2,"-o",color="red",linewidth=2.0)
ax2.set_title("Sales data of a Facewash")
ax2.set_xticks(df["month_number"])
ax2.set_xlabel("Month Number")
ax2.set_ylabel("Sale units in number")
plt.show()

#end code
```

## Exercise Question 10: Read all product sales data and show it using the stack plot

The Stack plot should look like this.

![Image 10](images/e10.png)

```python
# Excercise 10

#Start Code 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#read data
df = pd.read_csv("data/company_sales_data.csv")
x=df["month_number"]
data=[]
data.append(df["facecream"])
data.append(df["facewash"])
data.append(df["toothpaste"])
data.append(df["bathingsoap"])
data.append(df["shampoo"])
data.append(df["moisturizer"])

#plot
colors=["purple","aqua","red","black","darkgreen","gold"]
for i in range(len(data)):
    plt.plot([],[],color=colors[i],label=data[i].name)
plt.stackplot(x,data[0],data[1],data[2],data[3],data[4],data[5],colors = colors)

#labels and legend
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All product sales data using stack plot')
plt.legend(loc="upper left")
plt.show()

#end code
```
